```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import cross_val_score

from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('titanic_old.csv', usecols = ['Age','Fare','Survived', 'Parch','SibSp']) #Use columns which are necessary
```

```python
df.head()
```

```python
df ['Family'] = df['Parch'] + df['SibSp'] #Integrate between two columns
```

```python
df.head()
```

```python
df.drop(columns = ['SibSp','Parch']) #Drop the used integrated columns
```

```python
df.head()
```

```python
df.drop(columns = ['SibSp','Parch'], inplace = True)
x = df.drop(columns = ['Survived'])
y = df['Survived']
```

```python
x.head()
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
#Use Algo without Binarization
clf = DecisionTreeClassifier()
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
np.mean(cross_val_score(DecisionTreeClassifier(),x,y,cv=10,scoring='accuracy'))
```

```python
#Using Binarization
from sklearn.preprocessing import Binarizer
```

```python
trf = ColumnTransformer([
    ('bin',Binarizer(copy=False),['Family'])
],remainder='passthrough')
```

```python
X_train_trf = trf.fit_transform(x_train)
X_test_trf = trf.transform(x_test)
```

```python
pd.DataFrame(X_train_trf,columns=['family','Age','Fare'])
```

```python
clf.fit(X_train_trf,y_train)
```

```python
y_pred2 = clf.predict(X_test_trf)
```

```python
accuracy_score(y_test, y_pred2)
```
