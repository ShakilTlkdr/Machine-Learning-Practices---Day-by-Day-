```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.compose import ColumnTransformer
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic_old.csv', usecols = ['Age','Fare','Survived'])
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True) #Drop the missing values
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
x = df.iloc[:,1:]
y = df.iloc[:,0] # (define the target value & input value)
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2) #Split the data set for train and test
```

```python
df.head(2)
```

```python
clf =  DecisionTreeClassifier() #Used decision trees classifier
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
np.mean(cross_val_score(DecisionTreeClassifier(),x,y,cv=10,scoring='accuracy'))
```

```python
kbin_age = KBinsDiscretizer(n_bins = 10, encode = 'ordinal', strategy= 'quantile')
kbin_fare = KBinsDiscretizer(n_bins = 10, encode = 'ordinal', strategy= 'quantile')
```

```python
trf = ColumnTransformer([
   ('first', kbin_age, [0]),
    ('second', kbin_fare, [1])
])
```

```python
x_train_trf = trf.fit_transform(x_train)
x_test_trf = trf.fit_transform(x_test)
```

```python
clf.fit(x_train_trf, y_train)
```

```python
y_trf_pred = clf.predict(x_test_trf)
```

```python
accuracy_score(y_test, y_trf_pred)
```

```python
np.mean(cross_val_score(DecisionTreeClassifier(),x,y,cv=10,scoring='accuracy'))
```
